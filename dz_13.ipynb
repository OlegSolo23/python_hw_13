```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
user_id_100 = ratings.groupby('userId').count().filter(items = ['movieId']).query('movieId >= 100')
user_id_100
```

```python
long_life = ratings.groupby('userId').agg({'timestamp': ['min', 'max']})
long_life['diff'] = long_life['timestamp'] ['max'] - long_life['timestamp'] ['min']
long_life.head()
```

```python
avg_life = user_id_100.join(long_life['diff'], how='left')
avg_life['diff'].mean()
```
